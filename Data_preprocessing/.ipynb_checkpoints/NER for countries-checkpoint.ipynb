```python
import pandas as pd
import pycountry
import re
```

```python
df = pd.read_csv("youtube_death_penalty_dataset.csv")

df.head()
```

```python
countries = [c.name for c in pycountry.countries]

aliases = {
    "usa": "United States",
    "us": "United States",
    "america": "United States",
    "uk": "United Kingdom",
    "england": "United Kingdom",
    "sg": "Singapore",
    "singapore": "Singapore",
    "korea": "South Korea",
    "north korea": "North Korea",
    "south korea": "South Korea",
    "vietnam": "Viet Nam",
}
```

```python
def find_countries(text):
    text = str(text).lower()
    found = []

    for key, value in aliases.items():
        if re.search(rf"\b{key}\b", text):
            found.append(value)

    for country in countries:
        if re.search(rf"\b{country.lower()}\b", text):
            found.append(country)

    return list(set(found))
```

```python
df["countries"] = df["comment"].apply(find_countries)
```

```python
df.head()
```

```python
df.to_csv("youtube_comments_with_countries.csv", index=False)
```
