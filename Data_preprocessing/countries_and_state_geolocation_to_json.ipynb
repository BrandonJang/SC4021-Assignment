```python
import pandas as pd
```

```python
df = pd.read_csv("countries_and_states.csv")
df.head()
```

```python
df.columns = df.columns.str.strip()

js_array = "export const countries = [\n"

for _, row in df.iterrows():
    js_array += f'  {{ name: "{row["name"]}", lat: {row["latitude"]}, lng: {row["longitude"]} }},\n'

js_array += "];"

with open("countries_and_states.js", "w", encoding="utf-8") as f:
    f.write(js_array)
```
